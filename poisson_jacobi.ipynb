```python
import numpy as np
import plotly.express as px
from scipy.stats import norm
N = 1000
M = 2*N + 1
xs = np.linspace(-1, 1,num=M)
ys = np.zeros(M)
dx = 1 / M
diver = -norm(loc=0, scale=0.0001).pdf(xs)
```

```python
px.line(x=xs, y=diver)
```

```python
def jacobi_iterate(i, arr):
    theo = ((arr[i + 1] + arr[i - 1]) - diver[i]*dx**2) / 2
    diff = theo - arr[i]
    return diff
newys = ys.copy()
```

```python
f = 1
for i in range(10000):
    newys[1:-1] += f*jacobi_iterate(np.arange(1, M - 1), newys)
fig = px.line(x=xs, y=newys)
# fig.add_scatter(x=xs, y=0.1*np.minimum(1/(10*xs)**2, 1))
fig
```
